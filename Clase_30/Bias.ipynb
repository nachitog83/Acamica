En la presente clase vamos a entrar un poco más en detalle y vamos a presentar al `BIAS` asociado a un perceptrón.

Un valor de `BIAS` le permite al perceptrón mover la función de activación hacia la izquierda o hacia la derecha, lo que puede ser crítico para un aprendizaje exitoso. Miremos un ejemplo simple: considere esta red de 1 entrada y 1 salida que no tiene `BIAS`.

```python
from IPython.display import Image
Image("Imagenes/a.png")
```

La salida de la red se calcula multiplicando la entrada $x$ por el peso $w_0$ y pasando el resultado a través de algún tipo de función de activación (por ejemplo, una función sigmoide ($sig(x)$).

Aquí está la función que esta red calcula, para varios valores de $w_0$:

```python
from IPython.display import Image
Image("Imagenes/b.png")
```

Cambiar el peso $w_0$ esencialmente cambia la "inclinación" del sigmoide. Eso es útil, pero ¿y si quisieras que la red produjera $0$ cuando $x=2$? Simplemente cambiar la inclinación del sigmoide realmente no funcionará: desea poder desplazar toda la curva hacia la derecha.

Eso es exactamente lo que el `BIAS` te permite hacer. Si agregamos un `BIAS` al perceptrón anterior, llegamos al siguiente modelo:

```python
from IPython.display import Image
Image("Imagenes/c.png")
```

Entonces la salida de la red se convierte en $sig(w_0 * x + w_1 * 1.0)$. Así es como se ve la salida de la red para varios valores de $w_1$:

```python
from IPython.display import Image
Image("Imagenes/d.png")
```

### Tener un peso de $-5$ para $w_1$ desplaza la curva hacia la derecha, lo que nos permite tener una red que genera $0$ cuando $x=2$.
