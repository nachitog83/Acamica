```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
# La siguiente l√≠nea es para ver las imagenes dentro del notebook
%matplotlib inline
```

```python
felicidad = pd.read_csv('2016.csv',delimiter=';')

print('Dataset Felicidad: \n')
felicidad
```

```python
print('Dataset Felicidad Info: \n')
felicidad.info()
```

```python
print('Dataset Felicidad Describe: \n')

felicidad.describe()
```

```python
print('Cantidad de Pa√≠ses por Regi√≥n: \n')
felicidad.Region.value_counts()
```

```python
print('Correlaci√≥n entre valores del Dataset Felicidad: \n')
felicidad.corr()
```

```python
print('Correlaci√≥n entre valores del Dataset Felicidad: \n')
fe=felicidad.loc[:,['HappinessScore','Family','Economy','Health','Freedom']]
fecorr= fe.corr()
fecorr
```

```python
print('Gr√°fico Correlaci√≥n entre valores del Dataset Felicidad: \n')

plt.rcParams['figure.figsize'] = (12,10)
# Seteamos opciones de pandas sobre las columnas y su ancho
pd.set_option('max_columns', 120)
pd.set_option('max_colwidth', 5000)

sns.heatmap(fecorr, xticklabels=fecorr.columns, yticklabels=fecorr.columns);
```

```python

print('Promedio de Happiness Score por Regi√≥n: \n')

f= felicidad.groupby('Region')['HappinessScore'].mean().reset_index()

print(str(f.HappinessScore.sum()))

f.sort_values(by='HappinessScore')
```

```python
masFeliz = f.Region.iloc[f.sort_values(by='HappinessScore').tail(1).index].values

print('La region m√°s feliz considerando el promedio de HS de sus paises es: ' + str(masFeliz))
```

```python
menosFeliz = f.Region.iloc[f.sort_values(by='HappinessScore').head(1).index].values

print('La region menos feliz considerando el promedio de HS de sus paises es: ' + str(menosFeliz))
```

```python
print('Gr√°fico de barras Happiness Score por Regi√≥n: \n')

f.sort_values(by='HappinessScore',ascending=False,inplace=True)

plt.rcParams['figure.figsize'] = (12,8)


pd.set_option('max_columns', 120)
pd.set_option('max_colwidth', 5000)

a=sns.barplot(x='Region', y='HappinessScore',palette='pastel',data=f)
a.set_xlabel("Region",fontsize=16)
a.set_ylabel("Happiness Score",fontsize=16)
a.set_title("Promedio de Happiness Score por Region \n", fontsize=18)
plt.xticks(rotation=65);

```

```python
print('Gr√°fico de Torta % de HS por Regi√≥n: \n')

# Ac√° configuramos el tama√±o de las figuras
plt.rcParams['figure.figsize'] = (12,8)
# Seteamos opciones de pandas sobre las columnas y su ancho
pd.set_option('max_columns', 120)
pd.set_option('max_colwidth', 5000)

f.plot(x='Region',y='HappinessScore', kind='pie',startangle=20,autopct='%1.2f%%',colormap='Pastel1',title='Porcentaje de HS por Regi√≥n');
plt.legend(f['Region'],title='Happiness Score por Regi√≥n \n',bbox_to_anchor=(1.10, 0.5));

#loc='lower right',
#g.ax.legend(loc='center right', bbox_to_anchor=(1.00, 0.3), ncol=1)
```

```python
print('Gr√°fico de Econom√≠a de los pa√≠ses de la Regi√≥n Australia and New Zealand: \n')

plt.rcParams['figure.figsize'] = (8,6)

fe_Aus_Nz=felicidad[felicidad.Region == 'Australia and New Zealand'].copy()
fe_Aus_Nz.sort_values(by='Economy',ascending=False,inplace=True)

fe_Aus_Nz.plot(x='Country',y='Economy', kind='bar',title='Economy Region: Australia and New Zealand',grid=True,colormap='Pastel1')
plt.legend(fe_Aus_Nz['Country'],loc='lower right',title='Economy Region: Australia and New Zealand  \n')
plt.xticks(rotation='horizontal');
```

```python
print('Gr√°fico de Econom√≠a de los pa√≠ses de la Regi√≥n North America: \n')

plt.rcParams['figure.figsize'] = (8,6)


fe_Na=felicidad[felicidad.Region == 'North America'].copy()
fe_Na.sort_values(by='Economy',ascending=False,inplace=True)


fe_Na.plot(x='Country',y='Economy', kind='bar',title='Economy Region: North America',grid=True,colormap='Pastel1')
plt.legend(fe_Na['Country'],loc='lower right',title='Economy Region: North America  \n')
plt.xticks(rotation='horizontal');
```

```python
print('Gr√°fico de Econom√≠a de los pa√≠ses de la Regi√≥n Latin America and Caribbean: \n')

plt.rcParams['figure.figsize'] = (12,8)


fe_La=felicidad[felicidad.Region == 'Latin America and Caribbean'].copy()
fe_La.sort_values(by='Economy',ascending=False,inplace=True)


fe_La.plot(x='Country',y='Economy', kind='bar',title='Economy Region: Latin America and Caribbean',grid=True,colormap='Pastel1')
plt.legend(fe_La['Country'],loc='lower right',title='Economy Region: Latin America and Caribbean  \n')
plt.xticks(rotation='vertical');
```

```python
print('Gr√°fico de Familia de los pa√≠ses de la Regi√≥n Latin America and Caribbean: \n')

plt.rcParams['figure.figsize'] = (12,8)

fe_La.sort_values(by='Family',ascending=False,inplace=True)


fe_La.plot(x='Country',y='Family', kind='bar',title='Family Region: Latin America and Caribbean',grid=True,colormap='Pastel2')
plt.legend(fe_La['Country'],loc='lower right',title='Family Region: Latin America and Caribbean  \n'); 
```

```python
felicidadMaxima=felicidad[felicidad['HappinessScore']==felicidad.HappinessScore.max()]
print('Valores del Pa√≠s con mayor Felicidad vs mayores Valores: \n')
print('Economy: '+ str(felicidadMaxima.Economy.values) + ' Mayor valor:  '+ str(felicidad.Economy.max()))
print('Family: '+ str(felicidadMaxima.Family.values) + ' Mayor valor:  '+ str(felicidad.Family.max()) )
print('Health: '+ str(felicidadMaxima.Health.values) + ' Mayor valor:  '+ str(felicidad.Health.max()))
print('Freedom: '+ str(felicidadMaxima.Freedom.values) + ' Mayor valor:  '+ str(felicidad.Freedom.max()))



```

```python
#TOP 5 de valores de mayor econom√≠a
print('Mayores 5 valores de Econom√≠a: \n')

fel=felicidad.iloc[felicidad.Economy.sort_values(ascending=False).head(5).index]
fel.loc[:,['Region','Country','Economy']]

```

```python
print('Gr√°fico de Econom√≠a/Happiness Score por Pa√≠s \n')

plt.rcParams['figure.figsize'] = (11,7)

ax = sns.scatterplot(x="Economy", y="HappinessScore", hue="Country" , data=fe_La)
ax.legend(bbox_to_anchor=(1.30, 1.0));

```

```python
print('Ranking de Pa√≠ses m√°s felices, Regi√≥n: Latin America and Caribbean \n')
reg=['Latin America and Caribbean']

ordenFelicidad=felicidad[felicidad.Region.isin(reg)]
ordenFelicidad.sort_values(by='HappinessScore',ascending=False).head(10)
```

```python
print('Ranking de Felicidad de pa√≠ses con la econom√≠a m√°s baja \n')
economiaBaja=felicidad[felicidad.Economy<5 & felicidad.Region.isin(reg)]
economiaBaja.sort_values(by='HappinessScore',ascending=False).head(10)
```

```python
print('Gr√°fico de HS/Econom√≠a de pa√≠ses con baja Econom√≠a \n')
ax = sns.scatterplot(x="Economy", y="HappinessScore", hue="Country" , data=economiaBaja)
ax.legend(loc='lower right');
```
