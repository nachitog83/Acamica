# DBSCAN

El agrupamiento espacial basado en densidad de aplicaciones con ruido (DBSCAN) es un algoritmo de agrupación de datos basado en la densidad muy popular comúnmente utilizado en la minería de datos y el aprendizaje automático. 

DBSCAN agrupa los puntos de datos para separar las áreas de alta densidad de las áreas de baja densidad. También marca los puntos de datos como valores atípicos que se encuentran en las regiones de baja densidad. Los grupos formados pueden tener diferentes formas según la densidad de los puntos de datos.

## Idea básica del algoritmo DBSCAN

El algoritmo DBSCAN captura las regiones densas como grupos y requiere principalmente dos parámetros para encontrar grupos:

* `Epsilon:` la distancia máxima (distancia euclidiana) entre un par de puntos. Dos puntos se consideran vecinos si y solo si están separados por una distancia menor o igual a epsilon.
* `MinPoints:` la cantidad mínima de puntos necesarios para formar un grupo denso.

<img src='https://miro.medium.com/max/870/0*bUyZlx3rbNneiUA_'>

`MinPts = 4` significa que se requieren un mínimo de 4 puntos para formar un `grupo denso`. Además, un par de puntos deben estar separados por una distancia menor o igual a `Epsilon` para ser considerados vecinos. Según los dos parámetros anteriores, los puntos de datos se clasifican en 3 categorías de la siguiente manera:

* `Punto central:` Un punto seleccionado es un punto central si tiene al menos un número mínimo de puntos (MinPts) incluido en su vecindario  de distancia epsilon. En la figura anterior, los puntos rojos son puntos centrales que tienen al menos `MinPts = 4` en su vecindario. Si tenemos un punto central, significa que es una región densa.
* `Punto de borde:` Un punto seleccionado que está dentro de un vecindario de un punto central pero que en sí mismo no puede ser un punto central (menos de MinPts como vecinos). En la figura anterior, los puntos amarillos se identifican como puntos de borde. Si tenemos un punto de borde, significa que el punto está cerca o en el límite de una región densa.
* `Punto de ruido:` Un punto seleccionado que no es un punto central ni un punto de borde. Significa que estos puntos son valores atípicos que no están asociados con ningún grupo denso. En la figura anterior, el punto azul se identifica como punto de ruido.

## Algoritmo DBSCAN

1. Inicialmente, el algoritmo comienza seleccionando un `punto aleatorio` del conjunto de datos. Y comprueba si el punto seleccionado es un `punto central`. Una vez más, un punto es un punto central si contiene al menos el número `MinPoints` de puntos en su vecindario dentro de una distancia `epsilon`.

2. Luego, encuentra los `puntos conectados` a todos los `puntos centrales`, ignorando los `puntos no centrales`.

3. Asigna cada `punto no central` al clúster más cercano si el clúster se encuentra dentro de la distancia `epsilon` (`punto borde`). De lo contrario, lo asigna como `ruido`.

4. El algoritmo se detiene cuando explora todos los puntos uno por uno y los clasifica como `central`, `borde` y/o `ruido`.

# Implementación de DBSCAN usando Sklearn en Python

```python
# Importing DBSCAN model for clustering
from sklearn.cluster import DBSCAN
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_circles
```

```python
X, y = make_circles(n_samples=1000, factor=0.5, noise=.07, random_state=0)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x = X[:,0], y = X[:,1], hue = y)
plt.show()
```

## Instanciamos el modelo DBSCAN

* `Epsilon = 0.1`
* `MinPoints = 5`

```python
dbscan = DBSCAN(eps = 0.1, min_samples = 5)
```

## Entrenamos el modelo

```python
model = dbscan.fit(X)
```

## Mostramos los resultados

```python
# Funcion para plotear, NO CAMBIAR NADA
def plot_DBSCAN(X, labels, db, n_clusters):
    # Armamos una mascara, con unos en los datos que son CORES.
    core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
    core_samples_mask[db.core_sample_indices_] = True
    
    # Black removed and is used for noise instead.
    unique_labels = set(labels)
    colors = [plt.cm.Spectral(each)
              for each in np.linspace(0, 1, len(unique_labels))]
    plt.figure(figsize=(10,5))
    for k, col in zip(unique_labels, colors):
        if k == -1:
            # Black used for noise.
            col = [0, 0, 0, 1]
        class_member_mask = (labels == k)
        xy = X[class_member_mask & core_samples_mask]
        plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
                 markeredgecolor='k', markersize=14)
        xy = X[class_member_mask & ~core_samples_mask]
        plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
                 markeredgecolor='k', markersize=6)
    plt.title('Estimated number of clusters: %d' % n_clusters)
    plt.show()
    return
```

```python
labels = dbscan.labels_ #Etiquetas correspondientes a los clústeres que asigna el algoritmo
n_clusters = len(set(labels)) - (1 if -1 in labels else 0) #Número de clústeres que encontró el algoritmo
```

```python
plot_DBSCAN(X, labels, dbscan, n_clusters)
```
