# ¿Qué es la maldición de la dimensionalidad?

La maldición de la dimensionalidad se refiere a los fenómenos que ocurren al clasificar, organizar y analizar datos de `alta dimensión` que no ocurren en espacios de `baja dimensión`, específicamente el tema de la `escasez de datos` y la `cercanía de los datos.`

# Cuestiones
La escasez de datos se produce al pasar a dimensiones superiores. El `volumen del espacio representado crece tan rápido` que los datos no pueden mantenerse y, por lo tanto, se vuelven escasos. El problema de la escasez es importante para cualquier problema cuyo objetivo tenga cierta importancia estadística.

A medida que un espacio de datos se mueve de una dimensión a dos dimensiones y luego a tres dimensiones, los datos dados ocupan cada vez menos espacio de datos. Para mantener una representación precisa del espacio, los datos para el análisis crece exponencialmente.

El segundo problema que surge está relacionado con la clasificación de los datos. En espacios de dimensiones bajas, los datos pueden parecer muy similares, pero cuanto mayor es la dimensión, más lejos pueden parecer estos puntos de datos. Las dos turbinas eólicas a continuación parecen muy cercanas entre sí en dos dimensiones, pero separadas cuando se ven en una tercera dimensión. Este es el mismo efecto que la maldición de la dimensionalidad tiene en los datos.

# Características infinitas requieren entrenamiento infinito

Cuando se crean redes neuronales, se crean instancias con un cierto número de características (dimensiones). Cada dato tiene aspectos individuales, cada aspecto cae en algún lugar a lo largo de cada dimensión. En un ejemplo de clasificación de frutas, podemos querer una característica que maneje el color, una para el peso, una para la forma, etc. Cada característica agrega información, y si pudiéramos manejar todas las características posibles, podríamos saber perfectamente qué fruta estamos considerando. Sin embargo, una cantidad infinita de características requiere una cantidad infinita de ejemplos de entrenamiento, lo que elimina la utilidad del mundo real de nuestra red.

Lo más desconcertante es que la cantidad de datos de entrenamiento necesarios aumenta exponencialmente con cada característica agregada. Incluso si solo tuviéramos 15 características, cada una de las cuales es una pregunta de "sí" o "no" sobre el fruto que estamos identificando.

# Mitigando la maldición de la dimensionalidad

Una elección cuidadosa del número de dimensiones (características) que se utilizarán es la prerrogativa del científico de datos que capacita a una red. En general, cuanto más pequeño es el tamaño del conjunto de entrenamiento, menos funciones debe usar. Debe tener en cuenta que cada característica aumenta el requisito del conjunto de datos exponencialmente.

# Ejemplo en Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import seed
from numpy.random import randn
from mpl_toolkits.mplot3d import Axes3D
seed(42)
```

```python
# Genero los datos
i = 100 #cantidad de datos
sigma = 10
mu = 0
X = randn(i)*sigma+mu
y = randn(i)*sigma+mu
z = randn(i)*sigma+mu
```

# 1 Dimensión

```python
plt.figure(figsize=(10,5))
plt.scatter(X, [0]*i);
plt.show()
```

# 2 Dimensiones

```python
plt.figure(figsize=(10,5))
plt.scatter(X, y);
plt.show()
```

# 3 Dimensiones

```python
fig = plt.figure(figsize=(10,5))
ax = Axes3D(fig)
ax.scatter(X, y, z,s=60);
plt.show()
```
