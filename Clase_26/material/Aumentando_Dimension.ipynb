```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits import mplot3d
```

```python
x1 = np.array([1 , 2, -1, -2,  -1, -2.5,  1,  2])
x2 = np.array([.5, 1, .3, .75, -1, -.75,-.8,-.95])
clases = ['r','r','g','g','r','r','g','g']
```

```python
# Plot
plt.scatter(x1, x2, color=clases)
plt.title('Datos originales')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid()
```

```python
# Aumentando la dimension de los datos
X_1 = x1**2
X_2 = x2**2
X_3 = np.sqrt(2)*x1*x2
```

```python
%matplotlib inline
plt.figure(figsize=(15,10))
ax = plt.axes(projection='3d')
plt.grid()
ax.scatter3D(X_1, X_2, X_3, c=clases);
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3');
```
