# ¿Qué es un modelo?

Representación abstracta, conceptual, gráfica (o visual), física o matemática, de fenómenos, sistemas o procesos a fin de analizarlos, describirlos, explicarlos, simularlos y predecirlos.

### Ejemplo:

Supongamos que quiero modelar mi comportamiento como ahorrista, sólo teniendo en cuenta mis ingresos. Donde:
* A = Ahorros
* I = Ingresos

## Modelo lineal básico:
Ahora bien, decidimos modelar nuestro ahorro utilizando una función lineal básica, teniendo como premisa que __"yo siempre ahorro la mitad de lo que gano"__.

$$A(I) = 0.5 I$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
I = np.linspace(0,100,1000) #Ingresos
A = 0.5*I #Modelo
```

```python
plt.figure(figsize=(15,7))
plt.plot(I, A, color='g', label='Modelo propuesto')
plt.title('Modelo lineal de Ahorro', size=20)
plt.xlabel('Ingresos', size=15)
plt.ylabel('Ahorro', size=15)
plt.axis([0, 100, 0, 50])
plt.legend()
plt.show()
```

```python
noise = np.random.normal(0, 2, I.shape) #Genero ruido
A_real = np.power(I, 0.75) + noise #Datos reales
```

```python
plt.figure(figsize=(15,7))
plt.scatter(I, A_real, color='r', label='Datos reales')
plt.title('Modelo lineal de Ahorro', size=20)
plt.xlabel('Ingresos', size=15)
plt.ylabel('Ahorro', size=15)
plt.axis([0, 100, 0, 50])
plt.legend()
plt.show()
```

```python
A_new = np.power(I, 0.75) #Nuevo modelo
```

```python
plt.figure(figsize=(15,7))
plt.plot(I, A, color='g', label='Modelo propuesto')
plt.scatter(I, A_real, color='r', label='Datos reales')
plt.plot(I, A_new, color='b', label='Modelo mejorado')
plt.title('Modelo lineal de Ahorro', size=20)
plt.xlabel('Ingresos', size=15)
plt.ylabel('Ahorro', size=15)
plt.axis([0, 100, 0, 50])
plt.legend()
plt.show()
```
