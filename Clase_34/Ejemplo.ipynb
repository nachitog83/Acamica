# CÃ¡lculo de TF, IDF y TF-IDF

$$TF = \frac{Numero\ de\ repeticiones\ de\ la\ palabra\ en\ un\ documento}{Numero\ de\ palabras\ en\ un\ documento}$$


$$IDF = \log\left( \frac{Numero\ de\ documentos}{Numero\ de\ documentos\ que\ contienen\ la\ palabra}\right)$$


$$TFIDF = TF * IDF$$

<img src='https://cdn-media-1.freecodecamp.org/images/1*q3qYevXqQOjJf6Pwdlx8Mw.png'>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1,1000,10000)
y = np.log(x)

plt.figure(figsize=(15,7))
plt.plot(x, y)
plt.axis([0, 1000,-2, max(y)])
plt.xlabel('x', fontsize=15)
plt.ylabel('log(y)', fontsize=15)
plt.title('FunciÃ³n Log()', fontsize=20)
plt.grid()
plt.show();
```
