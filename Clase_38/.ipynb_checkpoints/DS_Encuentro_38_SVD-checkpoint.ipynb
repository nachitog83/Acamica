# SVD
### Ejemplo compresión de una imagen
Vamos a empezar con un ejemplo simple. SVD puede ser usado en el contexto de imágenes para comprimir fotos. Si en lugar de guardar la foto entera (Matriz `M` de la clase), uno guardase una representación reducida de esta matriz (`B`) y la manera de pasar de un espacio a otro (`V*`), entonces uno se podría ahorrar mucho espacio en el almacenado de una imagen. Este mismo proceso se utiliza a veces en imágenes y señales para reducir el ruido de las mismas (proceso llamado denoising)

EL formalismo de SVD nos asegura que dado el rango que eligamos (`r`), la foto que podamos reconstruir a partir de `B` y `V*` (`M aproximida`) será lo `mas parecida` posible a la foto original `M.`

`Primero cargamos la foto. Si no la tienen, deberan bajarse la libreria PIL para hacer este proceso mas fácil.`

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#plt.style.use('classic')
img = Image.open('domo.jpg')
print("Imagen original:")
plt.figure(figsize=(10, 5))
plt.imshow(img, cmap='gray')
plt.show()

# convert image to grayscale
imggray = img.convert('LA')

# convert to numpy array
imgmat = np.array(list(imggray.getdata(band=0)), float) # ver las dimensiones!

# Reshape according to orginal image dimensions
imgmat.shape = (imggray.size[1], imggray.size[0]) # por que aplicamos un reshape?

print("Imagen en escala de grises:")
plt.figure(figsize=(10, 5))
plt.imshow(imgmat, cmap='gray')
plt.show()
```

```python
imgmat.shape
```

`Vamos a usar la versión clasica del algoritmo de SVD que viene en numpy (noten qué importante que es, que viene por default en numpy, ya que se usa para muchas más cosas).`

```python
U, D, V = np.linalg.svd(imgmat)
```

`Inspeccionemos la dimensión de los elementos:`

```python
imgmat.shape
```

```python
U.shape
```

```python
D.shape
```

```python
V.shape
```

`Así como está, si multiplicamos adecuadamente estos tres elementos, recuperariamos la foto original. Veamos que si tomamos los 720 valores singulares (o sea todos) tendremos la imagen original nuevamente:`

```python
reconstimg = np.matrix(U[:, :720]) * np.diag(D[:720]) * np.matrix(V[:720, :])
print("Imagen en escala de grises reconstruida:")
plt.figure(figsize=(10, 5))
plt.imshow(reconstimg, cmap='gray')
plt.show()
```

---

`El objetivo pasa por tomar unicamente los valores singulares más altos. Por una cuestión de comodidad, los valores en D (notar que representa una matriz diagonal pero viene en forma de vector!) ya están ordenados de menor a mayor, siendo los primeros los de mayor valor. Ahora sí, vamos a probar tomando menos valores singulares, empecemos por sólo 2.`

```python
reconstimg = np.matrix(U[:, :2]) * np.diag(D[:2]) * np.matrix(V[:2, :])
print("Imagen en escala de grises con 2 valores singulares:")
plt.figure(figsize=(10, 5))
plt.imshow(reconstimg, cmap='gray')
plt.show()
```

```python
reconstimg.reshape(reconstimg.size)
```

`No es muy sorprendente el resultado, pero ya se puede al menos ver que hay un objeto en el centro de la imagen, para tener tan poca información (sólo los dos vectores propios asociados a los valores propios y los valores propios) no está tan mal. Ahora si, barramos una cantidad más razonable de valores singulares, y observemos los resultados.`

```python
A = reconstimg.reshape((reconstimg.size,))
A.shape
```

```python
for i in [5, 10, 15, 20, 30, 50,100]:
    reconstimg = np.matrix(U[:, :i]) * np.diag(D[:i]) * np.matrix(V[:i, :])
    print("Imagen en escala de grises con %d valores singulares:"%i)
    plt.figure(figsize=(10, 5))
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```

## Ejemplo con muchas imagenes

`Vamos a trabajar ahora con una serie de imagenes de caras de personas famosas. Nustro objetivo será encontrar un numero r de "caras base" (o "eigenfaces"). Estas son las mejores caras que, al ser combinadas de alguna manera en particular, pueden generar de vuelta todas las caras de nuestro dataset.`

```python
import os
import re
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
dir_data_caras ='lfw1000'

# Vemos cuantos archivos hay en la carpeta
print(len(os.listdir(dir_data_caras)))
```

`Buscamos los nombres de todos los archivos en la carpeta donde tenemos el dataset:`

```python
lista_data_caras = os.listdir(dir_data_caras)
lista_data_caras[:10]
```

`Definimos una funcion para leer este tipo de archivos:`

```python
def read_pgm(filename, byteorder='>'):
    """Return image data from a raw PGM file as numpy array.

    Format specification: http://netpbm.sourceforge.net/doc/pgm.html

    """
    with open(filename, 'rb') as f:
        buffer = f.read()
    try:
        header, width, height, maxval = re.search(
            b"(^P5\s(?:\s*#.*[\r\n])*"
            b"(\d+)\s(?:\s*#.*[\r\n])*"
            b"(\d+)\s(?:\s*#.*[\r\n])*"
            b"(\d+)\s(?:\s*#.*[\r\n]\s)*)", buffer).groups()
    except AttributeError:
        raise ValueError("Not a raw PGM file: '%s'" % filename)
    return np.frombuffer(buffer,
                            dtype='u1' if int(maxval) < 256 else byteorder+'u2',
                            count=int(width)*int(height),
                            offset=len(header)
                            ).reshape((int(height), int(width)))
```

`Para graficar las imagenes las debemos leer con la función que creamos y reproducirla con imshow de matplotlib. Notemos que cada imagen son 64 x 64 pixeles, en escala de grises.`

```python
image = read_pgm(dir_data_caras+'/'+lista_data_caras[0], byteorder='<')
plt.figure(figsize=(10, 5))
plt.imshow(image, plt.cm.gray)
plt.show()
image.shape
```

`Ahora vamos a armar una lista vacia y agregarle todas matrices de las imagenes dentro.`

```python
lista_matrices_imagenes = []
# Notar que recorremos todos los archivos de la carpeta y los leemos uno por uno.
for nombre in lista_data_caras:
    image = read_pgm(dir_data_caras+'/'+nombre, byteorder='<')
    lista_matrices_imagenes.append(image)
    #print(nombre)
    if image.shape != (64,64):
        print('La imagen no tiene 64x64')
```

`Llevamos cada imagen a un vector, y armamos una matriz donde el numero de filas es la cantidad de imagenes y las columnas son los 64x64 pixeles de cada imagen.`

```python
m = len(lista_matrices_imagenes)
d = image.shape[0] * image.shape[1]
X = np.reshape(lista_matrices_imagenes, (m, d))
```

`Para volver a pasar de fila de la matrix X a imagen solo debemos hacer el proceso inverso al reshape anterior.`

```python
# To get back to an image, just reshape it again
plt.figure(figsize=(10, 5))
plt.imshow((np.reshape(X[int(len(X)/2), :], (image.shape[0], image.shape[1]))), plt.cm.gray);
#plt.imshow((np.reshape(X[50, :], (image.shape[0], image.shape[1]))), plt.cm.gray)
```

`Ahora si, aplicamos el algoritmo de SVD a toda esta matriz. Notemos que se lo apicamos a todas las imagenes juntas al mismo tiempo. Esto es diferente a lo que hicimos en el ejercicio anterior, buscamos "caras base" que representen de la mejor manera posible a nuestra poblacion de caras.`

```python
U, D, V = np.linalg.svd(X, full_matrices=False)
```

```python
# Chequeamos dimensiones
print("X:    ", X.shape)
print("U:    ", U.shape)
print("Sigma:", D.shape)
print("VT:   ", V.shape)
```

`Ahora reconstruimos la matrix X con solo 100 valores principales`

```python
numero = 100
new_X = np.matrix(U[:, :numero]) * np.diag(D[:numero]) * np.matrix(V[:numero, :])
```

`Si volvemos a ver la cara anterior, vemos que ya podemos distinguirla, pero la resolución no es la optima. Prueben modificar el número de valores princpales hasta encontrar uno que les parezca adecuado.`

```python
plt.figure(figsize=(10, 5))
plt.imshow((np.reshape(new_X[int(len(new_X)/2), :], (image.shape[0], image.shape[1]))), plt.cm.gray);
#plt.imshow((np.reshape(new_X[50, :], (image.shape[0], image.shape[1]))), plt.cm.gray)
```

`Ahora si, podemos explorar cuales son las "caras base". Recordemos que nuestro "diccionario" para volver del espacio de la matriz B al espacio de la matriz original M, es la matriz de vectores singulares por derecha V*. Veamos que V* tiene la forma numero de instancias (1000) x cantidad de pixeles (64x64):`

```python
V.shape
```

`Las filas de V son entonces las vectores singulares por derecha, ordenados por orden de relevancia. V[0] es el que esta asociado al valor singular mas alto, V[1] al segundo, y así sucesivamente. Sumandolos de diferentes maneras los primeros r (por ejemplo, como vimos, los primeros 100), uno puede recuperar las imágenes originales. La infomación de como se deben sumar esta en la matriz U.`

Recordar que podemos reconstruir una $A'$ aproximada de la matriz $A$ original:

$$
A' = U  \Sigma_k V_k^T
$$

```python
for i in range(10,numero+10,10):
    print("Vector singular por derecha V[%d]:"%i)
    plt.figure(figsize=(10, 5))
    plt.imshow((np.reshape(V[i], (image.shape[0], image.shape[1]))), plt.cm.gray);
    plt.show()
```
