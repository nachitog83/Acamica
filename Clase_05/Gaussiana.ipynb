# Relación entre estadísticos y parámetros

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
np.random.seed(10) 
```

```python
mu, sigma = 5, 3 # mean and standard deviation
pocas_muestras = 5
muchas_muestras = 100000
s = np.random.normal(mu, sigma, pocas_muestras)
s2 = np.random.normal(mu, sigma, muchas_muestras)
```

```python
plt.figure(figsize = (10,8))
plt.hist(s, bins = 200, alpha = 1, label = 'Pocas muestras')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.hist(s2, bins = 200, alpha = 1, label = 'Muchas muestras')
plt.legend()
plt.grid()
plt.show()
```
