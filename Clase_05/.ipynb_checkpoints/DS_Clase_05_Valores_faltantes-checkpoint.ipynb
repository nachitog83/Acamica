````python
iris = pd.read_csv('DS_Clase_05_Iris_con_faltantes.csv')
iris_drop = pd.read_csv('DS_Clase_05_Iris_con_faltantes.csv') #Lo vamos a usar para eliminar valores faltantes
iris_fill = pd.read_csv('DS_Clase_05_Iris_con_faltantes.csv') #Lo vamos a usar para llenar valores faltantes
iris
````

## Funciones de Pandas para trabajar con valores faltantes
* ```python 
df.isna(), df.isnull()
```
* ```python
df.dropna(), df.dropna(subset=...)
```
* ```python
df.fillna(), df.fillna(inplace=...)
```
