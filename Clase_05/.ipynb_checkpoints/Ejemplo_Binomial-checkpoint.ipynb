# Variable Aleatoria Discreta con Distribución Binomial

Si tiro $n$ veces una moneda con probabilidad $p$ de sacar __CARA__. ¿Cuál es la probabilidad de sacar $x$ __CARAS__?

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
n = 10 #Cantidad de veces que tiro la moneda
p = 0.5 #Probabilidad de sacar cara
x = np.arange(n) #x caras
```

```python
fmp = stats.binom.pmf(x, n, p) #Distribución Binomial
plt.figure(figsize=(15,5))
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución de Probabilidad Binomial',fontsize=20)
plt.ylabel('Probabilidad',fontsize=15)
plt.xlabel('Valores',fontsize=15)
plt.show()
```
