# Reducción de dimensionalidad por convolución

Como pudimos observar a la salida de la convolución de una matriz con un filtro, obtenemos una nueva matriz de menor dimensión.

La fórmula general para conocer la dimensionalidad de salida dada una `matriz`, un `filtro`, `padding` y `stride`, es la siguiente:

$$n_{out}=\frac{n_{in}+2p-f}{s}+1$$

Donde $n_{out}$ es la dimensión de la matriz resultante (las matrices siempre deben ser cuadradas). Con $n_{in}$, como la dimensión de la matriz original. Con $p$, como el padding utilizado, además $f$ es la dimensión del filtro (también cuadrado) y $s$ el stride utilizado (normalmente se utiliza un stride de 1).

---
## Ejemplo:
Tenemos una imagen de `6x6` pixeles, un filtro de `3x3`, stride de 1 (salto de a 1 píxel) y no agrego padding, o sea:

* $n_{in}=6$
* $f=3$
* $p=0$
* $s=1$

Entonces $n_{out}=\frac{6+2*0-3}{1}+1=4$
