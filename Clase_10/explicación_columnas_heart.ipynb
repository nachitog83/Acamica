#### <b style="color: blue; font-size: 20px">PATIENT_ID</b> Identificador del paciente

#### <b style="color: blue; font-size: 20px">SLOPE_OF_PEAK_EXERCISE_ST_SEGMENT</b> 1: elevacion (upsloping), 2: horizontal,  3: depresion (downsloping). La Alteración del Segmento ST significa que existe un problema en la Repolarizacion Ventricular. Lo que se manifiesta en el Electrocardiograma como una Elevación o Depresión del Segmento ST.

```python
Image(filename= 'st.png', width=1000, height=300)
```

#### Principales causas de Elevación:

- Pericarditis aguda
- Hiperpotasemia
- Hipotermia.
- Síndrome de Brugada.
- Aneurisma ventricular
- Asociada a cambios de Repolarizacion (Repolarizacion Precoz, Bloqueos de rama, Marcapasos, Wolff-Parkinson-White.)

#### Principales causas de Depresión:

- Impregnación digitálica
- Taquicardia paroxística
- Asociado a cambios de Repolarizacion
- Esfuerzo Físico
- Hipopotasemia
- Hipertrofia ventricular izquierda
- Prolapso mitral

#### <b style="color: blue; font-size: 20px">THAL</b> Desorden sanguineo llamado talasemia (3 = normal; 6 = defecto fijo; 7 = defecto reversible) (La talasemia es un trastorno hereditario de la sangre. Se caracteriza por una cantidad inferior a la normal de las proteínas encargadas de transportar oxígeno (hemoglobina) y de glóbulos rojos. Los síntomas incluyen fatiga, debilidad, palidez y crecimiento lento)

#### <b style="color: blue; font-size: 20px">RESTING_BLOOD_PRESSURE</b> La presión arterial máxima de la persona (mm Hg - milímetros de mercurio - al ingreso al hospital)

#### <b style="color: blue; font-size: 20px">CHEST_PAIN_TYPE</b> El dolor torácico experimentado (Valor 1: angina típica, Valor 2: angina atípica, Valor 3: dolor no anginal, Valor 4: asintomática)

#### <b style="color: blue; font-size: 20px">NUM_MAJOR_VESSELS</b> Número de arterias coronarias principales coloreados por fluoroscopia (Fluoroscopia = Es un procedimiento en el que se utiliza un tinte especial (material de contraste) y rayos X para observar la forma en que fluye la sangre a través de las arterias en el corazón.)

#### <b style="color: blue; font-size: 20px">fasting_blood_sugar_gt_120_mg_per_dl</b> glucosa en sangre ( > 120 mg / dl, 1 = verdadero; 0 = falso) +120 prediabetis / diabetes

#### <b style="color: blue; font-size: 20px">resting_ekg_results</b> Medición electrocardiográfica en reposo (0 = normal, 1 = anormalidad en la onda ST-T, 2 = probable o definitiva hipertrofia ventricular izquierda)

#### <b style="color: blue; font-size: 20px">serum_cholesterol_mg_per_dl</b> El colesterol es una sustancia suave y cerosa que está presente en todo el cuerpo. Su cuerpo necesita un poco de colesterol para funcionar correctamente. Pero un exceso de colesterol puede tapar sus arterias y provocar enfermedades cardíacas. Normal menos de 200 mg/dl, Normal | Alto entre 200 y 240 mg/dl, Alto mas de 240 mg/dl

#### <b style="color: blue; font-size: 20px">oldpeak_eq_st_depression</b> Depresión del ST inducida por el ejercicio relativo al descanso

#### <b style="color: blue; font-size: 20px">sex</b> Sexo

#### <b style="color: blue; font-size: 20px">age</b> Edad

#### <b style="color: blue; font-size: 20px">max_heart_rate_achieved</b> La frecuencia cardíaca máxima alcanzada de la persona

#### <b style="color: blue; font-size: 20px">exercise_induced_angina</b> La angina de pecho es un tipo de dolor provocado por la reducción del flujo sanguíneo al corazón.  (1 = si; 0 = no)

#### <b style="color: blue; font-size: 20px">heart_disease_present</b> Enfermedad cardíaca presente
