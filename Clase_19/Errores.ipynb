```python
import numpy as np
np.random.seed(42)

noise = np.random.normal(0,1,20)
x = np.arange(0, 20, 1).tolist()
y = np.arange(0, 20, 1).tolist()
y_noise = x + noise
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10,5))
plt.plot(x, y, label="Regresión", color='k')
plt.scatter(x, y_noise, label="Dato Real")
plt.xlabel("Feature X", fontsize=15)
plt.ylabel("Salida y", fontsize=15)
plt.title("Errores", fontsize=20)
plt.grid()
plt.legend();
```

## Definimos los distintos errores

```python
error           = np.sum(y_noise - y)
mean_error      = error/len(y_noise)
abs_error       = np.sum(np.abs(y_noise - y))
mean_abs_error  = abs_error/len(y_noise)
sqrt_error      = np.sum(np.power(y_noise - y, 2))
mean_sqrt_error = sqrt_error/len(y_noise)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, y_noise - y, 'r--', label="Error")
plt.axhline(error, color='r', linestyle='-.', label="Suma de Error")
plt.axhline(mean_error, color='r', label="Error Medio")
plt.xlabel("Feature X", fontsize=15)
plt.ylabel("Salida y", fontsize=15)
plt.title("Error", fontsize=20)
plt.grid()
plt.xlim([0, 19])
plt.legend();
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, np.abs(y_noise-y), 'm--', label="Error Absoluto")
plt.axhline(abs_error, color='m', linestyle='-.', label="Suma Error Absoluto")
plt.axhline(mean_abs_error, color='m', label="Error Absoluto Medio")

plt.xlabel("Feature X", fontsize=15)
plt.ylabel("Salida y", fontsize=15)
plt.title("Error Absoluto", fontsize=20)
plt.grid()
plt.xlim([0, 19])
plt.legend();
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, np.power(y_noise-y,2), 'b--', label="Error Cuadrado")
plt.axhline(sqrt_error, color='b', linestyle='-.', label="Suma de Error Cuadrado")
plt.axhline(mean_sqrt_error, color='b', label="Error Cuadrado Medio")

plt.xlabel("Feature X", fontsize=15)
plt.ylabel("Salida y", fontsize=15)
plt.title("Error Cuadrático", fontsize=20)
plt.grid()
plt.xlim([0, 19])
plt.legend();
```
