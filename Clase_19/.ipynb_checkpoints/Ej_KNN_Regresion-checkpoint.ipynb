# Ejemplo de Regresión con KNN

```python
# Datos ficticios
X = [[0], [1], [2], [3]]
y = [0, 0, 1, 1]
to_predict = 1.5
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10,5))
plt.scatter(X, y, label="Datos Entrenamiento")
plt.xlabel("Feature X", fontsize=15)
plt.ylabel("Salida y", fontsize=15)
plt.title("KNN Regresión", fontsize=20)
plt.grid()
plt.xlim([-0.1, 3.1])
plt.ylim([-0.1, 1.1])
plt.vlines(to_predict, -0.1, 3.1, color='r', label="Valor a predecir")
plt.legend();
```

```python
from sklearn.neighbors import KNeighborsRegressor

reg = KNeighborsRegressor(n_neighbors=2)
reg.fit(X, y) 
prediccion = reg.predict([[1.5]])
print("Predicción: "+str(prediccion))
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X, y, label="Datos Entrenamiento")
plt.xlabel("Feature X", fontsize=15)
plt.ylabel("Salida y", fontsize=15)
plt.title("KNN Regresión", fontsize=20)
plt.grid()
plt.xlim([-0.1, 3.1])
plt.ylim([-0.1, 1.1])
plt.vlines(to_predict, -0.1, 3.1, color='r', label="Valor a predecir")
plt.scatter(to_predict, prediccion, color='m', marker='s', label="Predicción")
plt.legend();
```
