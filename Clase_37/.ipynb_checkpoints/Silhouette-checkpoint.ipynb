# Coeficiente de silueta: una mejor métrica

El coeficiente de silueta se calcula utilizando la `distancia media dentro del clúster` $a(i)$ y la `distancia media al clúster más cercano` $b(i)$, para cada muestra $i$. 

El coeficiente de silueta para una muestra $i$ es:

$$s(i)=\begin{cases}1-\frac{a(i)}{b(i)},&{\mbox{si }}a(i)<b(i)\\0,&{\mbox{si }}a(i)=b(i)\\\frac{b(i)}{a(i)}-1,&{\mbox{si }}a(i)>b(i)\\\end{cases}$$

Donde:
$$-1\leq s(i)\leq 1$$

Para aclarar, $b(i)$ es la distancia entre una muestra y el clúster más cercano del cual la muestra no forma parte. Podemos calcular el coeficiente de silueta medio sobre todas las muestras y usar esto como una métrica para juzgar el número de grupos.

# Vamos con un Ejemplo

Por ejemplo, generamos puntos de datos aleatorios utilizando la función `make_blob` de Scikit-learn con 4 features y 5 centros de clúster. 

La verdad fundamental del problema es que los datos se generan alrededor de 5 centroides. `Sin embargo, el algoritmo K-Means no tiene forma de saber esto.`

<img src='https://miro.medium.com/max/1133/1*589QI2NPYakYJty5i8m9kw.png'>

A continuación, ejecutamos el algoritmo de K-Means con una elección de $K=2, 3, 4, \dots , 11$ y calculamos el score de K-Means predeterminado (inercia) y el coeficiente de silueta promedio para cada uno de los valores de $K$ y los graficamos uno al lado del otro.

<img src='https://miro.medium.com/max/942/1*TxFPxOL-XJUsZ83Bnu31KA.png'>

La diferencia no podría ser más marcada. El coeficiente de silueta promedio aumenta hasta el punto cuando $K=5$ y luego disminuye bruscamente para valores más altos de $K$, es decir, exhibe un pico claro en $K=5$, que es el número de grupos con los que se generó el conjunto de datos original. El coeficiente de silueta exhibe una característica máxima en comparación con la curva suave en el método del codo, lo cual es es más fácil de visualizar y razonar.

Si aumentamos el ruido gaussiano en el proceso de generación de datos, los grupos se ven más superpuestos.

<img src='https://miro.medium.com/max/1124/1*bLymfNY9rcLOUtMDRwmO8g.png'>

En este caso, la puntuación K-Means predeterminada con el método de codo produce un resultado aún más ambiguo. En el diagrama de codo a continuación, es difícil elegir un punto adecuado donde se produce la curva real. ¿Es 4, 5, 6 o 7?

<img src='https://miro.medium.com/max/909/1*aI_dkLIlXW9EvpYjYcA8iQ.png'>

Pero la gráfica de coeficientes de silueta aún logra mantener una característica máxima alrededor de 4 o 5 centros de clúster.

De hecho, si miras las gráficas anteriores, se ve que en su mayoría hay 4 grupos visibles, aunque los datos se generaron utilizando 5 centros de grupos, debido a la alta varianza, solo se muestran estructuralmente 4 grupos. El coeficiente de silueta detecta este comportamiento fácilmente y muestra el número óptimo de grupos en algún lugar entre 4 y 5.

# En python

```python
# Preparamos una lista donde vamos a ir agregando los valores medios de silhouette
lista_sil = []
# Fiteammos un modelo para cada numero de cluster que queremos testear
for k in range(2,14):
    # Definimos y entrenamos el modelo
    km = KMeans(n_clusters=k)
    km = km.fit(X2)
    
    # Tomamos las etiquetas
    etiquetas = km.labels_
    
    # Calculamos el silhouette 
    valor_medio_sil = silhouette_score(X2, etiquetas)
    lista_sil.append(valor_medio_sil)
```

````python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
````

````python
# Generating the sample data from make_blobs
# This particular setting has one distinct cluster and 3 clusters placed close
# together.
X, y = make_blobs(n_samples=500,
                  n_features=2,
                  centers=4,
                  cluster_std=1,
                  center_box=(-10.0, 10.0),
                  shuffle=True,
                  random_state=1)  # For reproducibility

K = [2, 3, 4, 5, 6]
````

## Para tener en cuenta:

#### `Los coeficientes de silueta cerca de +1 indican que la muestra está muy lejos de los grupos vecinos. Un valor de 0 indica que la muestra está en o muy cerca del límite de decisión entre dos grupos vecinos y los valores negativos indican que esas muestras podrían haberse asignado al grupo incorrecto.`

````python
avg_sil = []

for n_clusters in K:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)
    
    avg_sil.append(silhouette_avg)
    
    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--", label='Avg. Silh. score')
    ax1.legend()

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(X[:, 0], X[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                c=colors, edgecolor='k')

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')
plt.show()
````

#### `En este ejemplo, el análisis de silueta se utiliza para elegir un valor óptimo para n_clusters. El gráfico de silueta muestra que el valor n_clusters de 3, 5 y 6 es una mala elección para los datos dados debido a la presencia de grupos con puntajes de silueta por debajo del promedio y también debido a amplias fluctuaciones en el tamaño de los gráficos de silueta. El análisis de silueta es más ambivalente al decidir entre 2 y 4.`

````python
plt.figure(figsize=(10,5))
plt.plot(K, avg_sil, 'b*--', label="Silhouette scores")
plt.scatter(4, avg_sil[2], color='r', marker='s', s=60, label='K elegido por Silhoutte')
plt.xlim(min(K),max(K))
plt.legend()
plt.grid()
plt.title("Silhouette");
plt.show()
````
